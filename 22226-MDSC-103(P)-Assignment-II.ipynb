Reddy-Mikke Problem   
$$
            Maximize: z = 5x_1+ 4x_2
$$
    subject to:   
$$
        6x_1 +4x_2 <= 24
$$
$$
        x_1+2x_2 <= 6
$$
$$
        -x_1+x_2 <= 1
$$
$$
        x_2 <= 2
$$
$$
        x_1,x_2 >= 0
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, Eq, solve
```

```python
x = np.linspace(0,5,100)
y = np.linspace(0,5,100)

z = lambda x,y: 5 * x + 4 * y

e1 = lambda x: (24 - 6*x)/4
e2 = lambda x: (6 - x)/2
e3 = lambda x: 1 + x
e4 = lambda x: 2 + x - x  

plt.figure(figsize = (10,5))
plt.xlabel('x', color='#1C2833', fontsize = 20)
plt.ylabel('y', color='#1C2833', fontsize = 20)
plt.plot(x, e1(x), label = "6*x + 4**y - 24")
plt.plot(x, e2(x), label = " x + 2*y - 6")
plt.plot(x, e3(x), label = " -x + y - 1")
plt.plot(x, e4(y), label = " y-2")

x,y = symbols('x y')
eq1 = Eq(x)
eq2 = Eq(y)
sol_dict = solve((eq1,eq2), (x,y), dict = True)
plt.plot(sol_dict[0][x], sol_dict[0][y], marker="o", c ="black")
plt.annotate("O", (sol_dict[0][x], sol_dict[0][y]), fontsize = 20)

x,y = symbols('x y')
eq1 = Eq(6 *x + 4*y - 24)
eq2 = Eq(y)
sol_dict = solve((eq1,eq2), (x,y), dict = True)
plt.plot(sol_dict[0][x], sol_dict[0][y], marker="o", c ="black")
plt.annotate("A", (sol_dict[0][x], sol_dict[0][y]), fontsize = 20)

x,y = symbols('x y')
eq1 = Eq(6 *x + 4*y - 24)
eq2 = Eq(x + 2*y -6)
sol_dict = solve((eq1,eq2), (x,y), dict = True)
plt.plot(sol_dict[0][x], sol_dict[0][y], marker="o", c ="green")
plt.annotate("B (Optimal)", (sol_dict[0][x], sol_dict[0][y]), fontsize = 20)

x,y = symbols('x y')
eq1 = Eq(-x + y -1)
eq2 = Eq(y-2)
sol_dict = solve((eq1,eq2), (x,y), dict = True)
plt.plot(sol_dict[0][x], sol_dict[0][y], marker="o", c ="black")
plt.annotate("C", (sol_dict[0][x], sol_dict[0][y]), fontsize = 20)

x,y = symbols('x y')
eq1 = Eq(x + 2*y -6)
eq2 = Eq(y-2)
sol_dict = solve((eq1,eq2), (x,y), dict = True)
plt.plot(sol_dict[0][x], sol_dict[0][y], marker="o", c ="black")
plt.annotate("D", (sol_dict[0][x], sol_dict[0][y]), fontsize = 20)

x,y = symbols('x y')
eq1 = Eq(-x + y -1)
eq2 = Eq(x)
sol_dict = solve((eq1,eq2), (x,y), dict = True)
plt.plot(sol_dict[0][x], sol_dict[0][y], marker="o", c ="black")
plt.annotate("E", (sol_dict[0][x], sol_dict[0][y]), fontsize = 20)

plt.axvline(x=0, c="black")
plt.axhline(y=0, c="black")

plt.legend()
plt.show
```
